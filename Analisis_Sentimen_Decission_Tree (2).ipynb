```python
#preprocessing
import re
def casefolding(comment):
  comment = comment.lower()
  comment = comment.strip(" ")
  comment = re.sub(r'[?|$|.|!Â²_:")(-+,]', '', comment)
  return comment

data['Comment'] = data['Comment'].apply(casefolding)
data.head(10)
```

```python
#tokenizing
def token(comment):
  nstr = comment.split(' ')
  dat= []
  a = -1
  for hu in nstr:
    a = a + 1
  if hu == '':
    dat.append(a)
  p = 0
  b = 0
  for q in dat:
    b = q - p
    del nstr[b]
    p = p + 1
  return nstr
data['Comment'] = data['Comment'].apply(token)
data.head(10)
```

```python
data = data.astype({'sentimen' : 'category'})
data = data.astype({'Comment' : 'string'})
data.dtypes
```

```python
#tf-idf
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()
text_tf = tf.fit_transform(data['Comment'].astype('U'))
text_tf
```

```python
#menghitung sentimen positif negatif
data['sentimen'].value_counts()
```

```python
#splitting
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(text_tf, data['sentimen'], test_size=0.2, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, confusion_matrix

model = DecisionTreeClassifier().fit(X_train, y_train)
prediksi = model.predict(X_test)
print('Accuracy ',accuracy_score(y_test, prediksi))
print('Precision ',precision_score(y_test, prediksi, average='weighted'))
print('Recall ',recall_score(y_test, prediksi, average='weighted'))
print('Confusion Matrix\n', confusion_matrix(y_test, prediksi))
print('Classification Report\n', classification_report(y_test, prediksi, zero_division=0))
```
